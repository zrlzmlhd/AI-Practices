```python
from sklearn.datasets import load_iris
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()

ada_clf = AdaBoostClassifier(DecisionTreeClassifier(max_depth= 1),n_estimators= 200, learning_rate= 0.5, random_state= 42,algorithm= "SAMME")

ada_clf.fit(iris["data"], iris["target"])
print(ada_clf.score(iris["data"], iris["target"]))
```
