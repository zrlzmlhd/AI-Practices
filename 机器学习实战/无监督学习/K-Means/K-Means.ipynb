```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
X, y = load_iris().data, load_iris().target

k= 5
kmeans = KMeans(n_clusters= k)
y_pred = kmeans.fit_predict(X)
print(y_pred)
# 分类的实例标签是该实例被分配给该集群的索引
print(y_pred is kmeans.labels_)

# 查看一下算法发现的5个算法中心点
kmeans.cluster_centers_
```

# 将新实例分配给离中心点最近的镞

```python
import numpy as np
X_new = np.array([[0,2,1,3], [3,4,5,6], [-3,5,4, 3], [5,2,3, 6]])
# 预测新实例所属的簇
y_new_predict = kmeans.predict(X_new)
```

# 为每个Kmeans采用软聚类 分数值为相似性分数 例如高斯径向基函数

```python
kmeans.transform(X_new)
```
