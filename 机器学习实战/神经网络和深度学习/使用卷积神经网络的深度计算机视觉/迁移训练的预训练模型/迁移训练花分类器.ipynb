```python
import tensorflow as tf

# 1. 查看TensorFlow版本
print("TensorFlow版本：", tf.__version__)

# 2. 检测GPU是否可用
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    print("GPU设备列表：", gpus)
    print("GPU可用：True")
else:
    print("GPU可用：False（若需GPU加速，需确认CUDA、cuDNN配置是否匹配）")

# 3. 测试基础张量计算
a = tf.constant([[1.0, 2.0], [3.0, 4.0]])
b = tf.constant([[5.0, 6.0], [7.0, 8.0]])
c = tf.matmul(a, b)
print("张量乘法结果：\n", c.numpy())
```

```python
import sys
import tensorflow as tf
import torch
import platform

print("=" * 60)
print(f"当前 Python 版本: {sys.version}")
print(f"当前操作系统: {platform.system()} {platform.release()}")
print("=" * 60)

# --- 1. TensorFlow GPU 检测与测试 ---
print("### TensorFlow (TF) GPU/CUDA 检测 ###")
print("-" * 30)

# 检查 TF 是否能找到可用的 GPU 设备
gpu_devices = tf.config.list_physical_devices('GPU')

if gpu_devices:
    print(f"✅ TF 找到了 {len(gpu_devices)} 个 GPU 设备:")
    for device in gpu_devices:
        print(f"   - {device}")

    # 简单测试：将一个张量放在 GPU 上进行运算
    try:
        with tf.device(gpu_devices[0].name):
            a = tf.constant([1.0, 2.0, 3.0], dtype=tf.float32)
            b = tf.constant([4.0, 5.0, 6.0], dtype=tf.float32)
            c = a + b
            print(f"   - TF 运算测试成功: {c.numpy()}")
            print(f"   - 张量设备: {c.device}")
            print("✅ TF GPU 成功检测并运行！")
    except Exception as e:
        print(f"❌ TF GPU 运算测试失败: {e}")
        print("请检查您的 CUDA/cuDNN 版本和 TF 的兼容性。")
else:
    print("❌ TF 没有找到可用的 GPU 设备。")
    print("请确认 GPU 驱动已安装，且 CUDA/cuDNN 版本与 TF 版本兼容。")

print("-" * 30)


# --- 2. PyTorch GPU 检测与测试 ---
print("\n### PyTorch GPU/CUDA 检测 ###")
print("-" * 30)

# 检查 PyTorch 是否支持 CUDA
if torch.cuda.is_available():
    cuda_count = torch.cuda.device_count()
    current_device_name = torch.cuda.get_device_name(0)

    print(f"✅ PyTorch 找到了 {cuda_count} 个 CUDA 设备。")
    print(f"   - 设备名称: {current_device_name}")

    # 简单测试：将一个张量放在 GPU 上进行运算
    try:
        device = torch.device("cuda:0")
        x = torch.tensor([1.0, 2.0, 3.0], device=device)
        y = torch.tensor([4.0, 5.0, 6.0], device=device)
        z = x * y
        print(f"   - PyTorch 运算测试成功: {z}")
        print(f"   - 张量设备: {z.device}")
        print("✅ PyTorch GPU 成功检测并运行！")

    except Exception as e:
        print(f"❌ PyTorch GPU 运算测试失败: {e}")
        print("请检查您的 CUDA/cuDNN 版本和 PyTorch 的兼容性。")
else:
    print("❌ PyTorch 找不到 CUDA 设备。")
    print("请确认 PyTorch 是安装的 CUDA 版本，而不是 CPU 版本。")

print("-" * 30)
```
