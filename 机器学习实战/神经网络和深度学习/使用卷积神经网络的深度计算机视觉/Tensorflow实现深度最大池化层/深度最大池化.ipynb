```python
import numpy as np

# 设置随机种子
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
import tensorflow as tf
from tensorflow import keras
from sklearn.datasets import load_sample_image
china = load_sample_image("china.jpg") / 255
flower = load_sample_image("flower.jpg") /255
images = np.array([china, flower], dtype=np.float32)
with tf.device('/CPU:0'): # 深度最大池化目前只能在CPU上面运行
    output = tf.nn.max_pool(images, ksize=(1,1,1,3),strides=(1,1,1,3),padding='SAME')
```

## 将这个深度最大池化层作为keras一个层

```python
depth_pool = keras.layers.Lambda(lambda X: tf.nn.max_pool(X, ksize=(1,1,1,3), strides=(1,1,1,3), padding='SAME'))
```

## 全局平均池化层


```python
global_pool = keras.layers.GlobalAveragePooling2D()
# 等效于
global_pool = keras.layers.Lambda(lambda X: tf.reduce_mean(X, axis=[1, 2]))
```
