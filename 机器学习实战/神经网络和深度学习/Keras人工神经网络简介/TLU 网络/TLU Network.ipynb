```python
import numpy as np

# 设置随机种子
np.random.seed(42)
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron

iris = load_iris()
X = iris.data[:, [2, 3]]
y = (iris.target == 0).astype(int)  # 仅分类Iris-setosa与非Iris-setosa
per_clf = Perceptron()
per_clf.fit(X, y)

y_pred = per_clf.predict([[2, 0.5]])
```
