# 张量可以和numpy结合使用

```python
import numpy as np

# 设置随机种子
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
import tensorflow as tf
t = tf.constant([2,5,6])
a = np.array([1,2,3])
tf.constant(a)
np.array(t)
```

```python
#  默认情况下numpy使用64位精度 tensorflow使用32位精度 所以使用numpy创建张量时要设置类型
a = np.array([1,2,3],dtype=np.float32)
```
