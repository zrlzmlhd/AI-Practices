```python
# 使用函数式API创建多输出模型
from keras import layers
from keras.layers import Input
from keras.models import Model

vocabulary_size = 50000
num_income_groups =10

posts_input = Input(shape=(None,), dtype="int32", mane= "posts")
embedded_posts = layers.Embedding(256, vocabulary_size, 256)(posts_input)
x = layers.Conv1D(123, 5, activation='relu')(embedded_posts)
x = layers.MaxPooling1D(5)(x)
x = layers.Conv1D(256, 5, activation='relu')(x)
x = layers.Conv1D(256, 5, activation='relu')(x)
x = layers.MaxPooling1D(5)(x)
x = layers.Conv1D(256, 5, activation='relu')(x)
x = layers.Conv1D(256, 5, activation='relu')(x)
x = layers.GlobalMaxPooling1D()(x)
x = layers.Dense(128, activation='relu')(x)

age_predictions = layers.Dense(1, name='age', activation='relu')(x) # 输出层都具有名称
income_predictions = layers.Dense(num_income_groups, name='income', activation='softmax')(x)
gender_predictions = layers.Dense(1, name='gender', activation='sigmoid')(x)

# 使用Input 和 Output 创建多输出模型
model = Model(posts_input, [age_predictions, income_predictions, gender_predictions])

model.compile(optimizer='adam',
              loss={'age': 'mse',
                    'income': 'categorical_crossentropy',
                    'gender': 'binary_crossentropy'},
              loss_weights={'age': 0.5, 'income': 1, 'gender': 10.0})

model.summary()

# 训练模型 这里我就不添加数据了 选择你的数据集
model.fit(posts, [age_targets, income_targets, gender_targets], epochs=10, batch_size=64)
```
