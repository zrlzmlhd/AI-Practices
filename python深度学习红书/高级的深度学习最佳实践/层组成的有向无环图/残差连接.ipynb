```python
#残差连接是为了解决深度神经网络在训练过程中出现的梯度消失和梯度爆炸问题。通过引入残差连接，网络可以更容易地学习到恒等映射，从而使得信息能够更有效地传递到更深层的网络中。
# 残差连接的基本思想是将输入直接添加到输出上，从而形成一个“捷径”路径，使得梯度可以直接传递到更浅层的网络中，避免了梯度消失的问题。
# 这种连接方式使得网络能够更容易地学习到恒等映射，从而使得信息能够更有效地传递到更深层的网络中。
# 残差连接的引入使得深度神经网络能够更容易地训练，并且在许多任务中取得了显著的性能提升。
```

```python
# 假设数据为x
from keras import layers
y = layers.Conv2D(256, 3, padding='same', activation='relu')(x)
y = layers.Conv2D(256, 3, padding='same', activation='relu')(y)
# 残差连接
y = layers.add([x, y])
```

```python
# 如果特征图尺寸不匹配 使用以下方法调整
y = layers.Conv2D(256, 3, padding='same', activation='relu')(x)
y = layers.Conv2D(256, 3, padding='same', activation='relu')(y)
y = layers.MaxPool2D(2, strides= 2)(y)

# 使用1x1卷积调整x的通道数和尺寸
x = layers.Conv2D(256, 1, padding='same', strides=2)(x)
# 残差连接
y = layers.add([x, y])
```
