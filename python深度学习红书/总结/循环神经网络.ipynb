## RNN
如果数据的时间顺序十分重要 那么应该优先是用RNN而不是CNN
## 三种RNN
1. Simple RNN
2. LSTM ：缺点是计算代价高
3. GRU ：可以理解为Lstm的轻量化版本 更加简单 计算代价更小

## RNN的链接
如果要进行多个RNN链接 那么要保证最后一层之前的每一层都要返回输出的完整序列

```python
## 单个RNN
from keras import models, layers

model = models.Sequential()
model.add(layers.LSTM(32, input_shape=(num_timesteps, num_features)))
model.add(layers.Dense(num_classes. activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
```

```python
## 多个RNN链接
model = models.Sequential()
model.add(layers.LSTM(32, return_sequences=True, input_shape=(num_timesteps, num_features)))
model.add(layers.LSTM(32, return_sequences=True))
model.add(layers.LSTM(32))
model.add(layers.Dense(num_classes, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
```
