## 密集层连接网络
1.由Dense层的堆叠组成
2.常用来进行分类数据任务
3.还用于大多数网络架构的最终分类或者回归阶段

## 二分类问题
1.使用sigmoid和一层的dense
2.layers.Dense(1, activation='sigmoid')

## 单标签多分类问题
1.使用softmax
2.单元个数等于类别个数
3.如果目标是one-hot编码的 那么loss函数应该使用categorical_crossentropy
4.如果目标是整数编码的 那么loss函数应该使用sparse_categorical_crossentropy

## 多标签分类问题
1.使用sigmoid和多个dense层
2.每个dense层对应一个类别
3.每个dense层都使用sigmoid激活函数
4.目标是k-hot编码的 那么loss函数应该使用binary_crossentropy
