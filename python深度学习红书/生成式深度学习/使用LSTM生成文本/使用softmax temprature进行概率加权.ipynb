```python
# 对不同的概率分布进行重新加权
import numpy as np

# 设置随机种子
np.random.seed(42)

def reweight_distribution(original_distribution, temperature=0.5):
    """
    对原始分布进行重新加权，以生成新的分布。
    :param original_distribution: 概率值组成的一维数组 求和为1
    :param temperature: 定量描述输出的熵
    :return: 重新加权后的分布
    """
    distribution = np.log(original_distribution) / temperature
    distribution = np.exp(distribution)
    # distribution的求和不再可能等于1, 因此需要将他除以求和 得到新的分布
    return distribution / np.sum(distribution)
```
